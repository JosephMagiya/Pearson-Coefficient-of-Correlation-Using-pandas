```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# Link to the dataset - https://www.kaggle.com/russellyates88/suicide-rates-overview-1985-to-2016

SuicideRate = pd.read_csv("suicide-rates-overview-1985-to-2016.csv")
SuicideRate.head()
```

```python
pearsoncorr = SuicideRate.corr(method='pearson')

# pearsoncorr
sb.heatmap(pearsoncorr, 
            xticklabels=pearsoncorr.columns,
            yticklabels=pearsoncorr.columns,
            cmap='RdBu_r',
            annot=True,
            linewidth=0.5)

# The darker - The higher the corelation
```
